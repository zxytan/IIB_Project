```python
from optimise_structure import optimise
import time
import numpy as np
import csv
from structure import structure
```

```python
#collect more data - trying to use less time to get better result
#do 10 opts - randomness can have effect
# choose - equiv_cant lf model, 20*n hf iter, 50*n lf iter
lf_model = "equiv_cant"
for i in range(10):
    print(i)
    for n in [1, 2, 3, 4, 5]:
        print(n)
        hf_max_iter = 20*n
        lf_max_iter = 50*n
        opt = optimise(n, "beam", lf_model, hf_max_iter, lf_max_iter)
        
        start = time.time()
        opt.run_hf_opt()
        end = time.time()
        hf_time = end-start
        
        lf_start = time.time()
        opt.run_lf_opt()
        end = time.time()
        lf_time = end-lf_start
        
        mf_start = time.time()
        opt.run_mf_opt()
        end = time.time()
        mf_time = end-mf_start
        total_time = end-start
        
        opt.report_opt_results()
        results = opt.results
        results['hf time'] = hf_time
        results['lf time'] = lf_time
        results['mf time'] = mf_time
        results['total time'] = total_time
        
        with open('n_opt.csv', 'a+') as output_file:
            fieldnames = results.keys()
            dict_writer = csv.DictWriter(output_file, dialect="excel", fieldnames=fieldnames)
            dict_writer.writerow(results)
```
